```python
import matplotlib.pyplot as plt
import numpy as np
years = [2004, 2006, 2008, 2010, 2012]
sachin_runs = [850, 900, 1200, 1500, 800]
kohli_runs = [0, 0, 500, 1000, 1100]
sehwag_runs = [1100, 1050, 1000, 900, 700]
bar_width = 0.25
x = np.arange(len(years))
plt.figure(figsize=(6, 4))
plt.bar(x - bar_width, sachin_runs, width=bar_width, label='Sachin')
plt.bar(x, kohli_runs, width=bar_width, label='Kohli')
plt.bar(x + bar_width, sehwag_runs, width=bar_width, label='Sehwag')

plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs by Year')
plt.xticks(x, years)
plt.legend()
plt.tight_layout()
plt.show()

players = ['Sachin', 'Kohli', 'Sehwag']
debut_runs = [4000, 4500, 4200]

plt.figure(figsize=(6, 2))
plt.barh(players, debut_runs, color=['blue', 'green', 'orange'])
plt.xlabel('Total Runs')
plt.title('Total Runs in Debut 5 Years')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Android', 'iOS', 'Others']
shares = [72, 26, 2]

plt.figure(figsize=(3, 3))
plt.pie(shares, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Mobile OS Market Share')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(6, 2))
plt.barh(labels, shares, color=['red', 'pink', 'yellow'])
plt.xlabel('Market Share (%)')
plt.title('Mobile OS Market Share')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['APPLE', 'BANANA', 'MANGO', 'PINEAPPLE', 'JACKFRUIT']
values = [16, 13, 30, 25, 5]

plt.bar(categories, values, color='skyblue')

for i, value in enumerate(values):
    plt.text(i, value + 0.5, str(value), ha='center')

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Annotations")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x), color='yellow', linestyle='-', label='sin(x)')
plt.plot(x, np.cos(x), color='grey', linestyle='--', label='cos(x)')
plt.plot(x, np.exp(x), color='red', linestyle=':', label='exp(x)')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines in One Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 50)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
scatter_x = np.random.rand(20)
scatter_y = np.random.rand(20)
hist_data = np.random.randn(1000)

fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title("Line Plot")

axs[0, 1].bar(categories, values, color='green')
axs[0, 1].set_title("Bar Chart")

axs[1, 0].scatter(scatter_x, scatter_y, color='red')
axs[1, 0].set_title("Scatter Plot")

axs[1, 1].hist(hist_data, bins=20, color='orange', edgecolor='black')
axs[1, 1].set_title("Histogram")

plt.suptitle("2x2 Grid of Subplots")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)
product_a = np.random.randint(5, 15, size=12)
product_b = np.random.randint(3, 10, size=12)
product_c = np.random.randint(2, 8, size=12)

plt.stackplot(months, product_a, product_b, product_c,
              labels=['Product A', 'Product B', 'Product C'],
              colors=['yellow', 'red', 'blue'])

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Cumulative Sales of Products Over 12 Months")
plt.legend(loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt

companies = ['TechNova', 'GreenEarth Energy', 'Skyline Airlines', 'FreshBite Foods']
market_share = [35, 25, 20, 20]

explode = [0.1 if share == max(market_share) else 0 for share in market_share]

plt.pie(market_share, labels=companies, autopct='%1.1f%%', explode=explode,
        shadow=True, startangle=90)

plt.title("Market Share of Companies")
plt.show()
```
